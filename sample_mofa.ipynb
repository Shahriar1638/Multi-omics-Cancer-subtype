```python
import pandas as pd

expression_df = pd.read_csv("processed_expression.csv")
methylation_df = pd.read_csv("processed_methylation(1).csv")



print("\nExpression DataFrame head:")
display(expression_df.head())

print("\nMethylation DataFrame head:")
display(methylation_df.head())
print("\nExpression DataFrame shape:", expression_df.shape)
print("Methylation DataFrame shape:", methylation_df.shape)
```

```python
# Create a dictionary of dataframes for MOFA+
mdata = {
    'gene_expression': expression_df,
    'methylation': methylation_df,
    # 'cnv': cnv_df  # Uncomment and define cnv_df if you have CNV data
}

# Display the keys and the shapes of the dataframes in the dictionary to verify
print("MOFA+ data dictionary keys and shapes:")
for key, df in mdata.items():
    print(f"{key}: {df.shape}")
```

```python
from mofapy2.run.entry_point import entry_point

# Prepare the data: remove non-numeric columns
if 'Ensembl_ID' in expression_df.columns:
    expression_numeric = expression_df.drop(columns=['Ensembl_ID'])
else:
    expression_numeric = expression_df.copy()

if 'Composite Element REF' in methylation_df.columns:
    methylation_numeric = methylation_df.drop(columns=['Composite Element REF'])
else:
    methylation_numeric = methylation_df.copy()

# Wrap into a nested list (1 group with 2 views)
data_list = [
    [expression_numeric.values, methylation_numeric.values]
]

# Initialize MOFA
ent = entry_point()

# Data options
ent.set_data_options(
    scale_groups=False,
    scale_views=False
)

# Set data matrix
ent.set_data_matrix(data_list)

# Model options
ent.set_model_options(
    factors=15,
    spikeslab_factors=True,
    ard_factors=True,
    ard_weights=True
)

# Training options
ent.set_train_options(
    iter=1000,
    convergence_mode="medium",
    seed=42,
    verbose=True
)

# Build + train
ent.build()
ent.run()

# Get trained model
model = ent.model
```

```python
# Factors (Z) and weights (W) are stored in the trained model
factors = ent.model.getFactorsArray()  # numpy array: samples x factors
weights = ent.model.getWeightsArray()  # numpy array: features x factors
```

```python
outfile = "/Users/ricard/data/mofaplus/hdf5/test.hdf5"

# - save_data: logical indicating whether to save the training data in the hdf5 file.
# this is useful for some downstream analysis in R, but it can take a lot of disk space.
ent.save(outfile, save_data=True)

#########################
## Downstream analysis ##
#########################

# Check the mofax package for the downstream analysis in Python: https://github.com/bioFAM/mofax
# Check the MOFA2 R package for the downstream analysis in R: https://www.bioconductor.org/packages/release/bioc/html/MOFA2.html
# All tutorials: https://biofam.github.io/MOFA2/tutorials.html

# Extract factor values (a list with one matrix per sample group)
factors = ent.model.nodes["Z"].getExpectation()

# Extract weights  (a list with one matrix per view)
weights = ent.model.nodes["W"].getExpectation()

# Extract variance explained values
r2 = ent.model.calculate_variance_explained()

# Interact directly with the hdf5 file
import h5py
f = h5py.File(outfile, 'r')
f.keys()

# Extract factors
f["expectations"]["Z"]["group_0"].value
f["expectations"]["Z"]["group_1"].value

# Extract weights
f["expectations"]["W"]["view_0"].value
f["expectations"]["W"]["view_1"].value

# Extract variance explained estimates
f["variance_explained"]["r2_per_factor"]
f["variance_explained"]["r2_total"]
```
